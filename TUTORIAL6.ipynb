```python
import numpy as np
import matplotlib.pyplot as plt 


x=np.array([0,1,2,3])
y=np.array([-1.0,0.2,0.9,2.1])
N=len(x)

sumx=np.sum(x)
sumy=np.sum(y)
sumxy=np.sum(x*y)
sumx2=np.sum(x**2)

m=(N*sumxy-sumx*sumy)/(N*sumx2-sumx**2)
b=(sumy-m*sumx)/N

print(f"Gradient: {m}")
print(f"Intercept: {b}")

y_p=m*x+b

plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue')  # Original data points
plt.plot(x, y_p, color='red', label=f'Best-fit line: y = {m:.2f}x + {b:.2f}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Least Squares Fitting')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

nobs=100

xobs=np.random.uniform(low=0.0, high=200.0, size=nobs)
noise=np.random.normal(loc=0.0, scale=5.0, size=nobs)

yobs=2*xobs+5+noise
x=xobs.reshape(-1,1)
y=yobs

lr=LinearRegression()
lr.fit(x,y)

print('The coefficient = ',lr.coef_)
print('The intercept =',lr.intercept_)

fig=plt.figure(figsize=(21,8))
fig.suptitle('Machine Learning Linear Regression in Supervised Learning - \Regression Algorithm.',color='m',fontsize=16)
plt.plot(xobs,yobs,'o')
xgrid=np.linspace(xobs.min(),xobs.max(),num=100)
plt.plot(xgrid, lr.coef_[0]*xgrid+lr.intercept_)

xpredict=np.array([50,100,150,200])
ypredict1=lr.intercept_+xpredict*lr.coef_[0]
ypredict2=lr.predict(xpredict.reshape(-1,1))

print('Predictions using the linear regression supervised learning algorithm')
print('Selected x values: ',xpredict)
print('Method 1: ',ypredict1)
print('Method 2: ',ypredict2)

plt.show()
plt.savefig('t6q6.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

nobs=100

xobs=np.random.uniform(low=0.0, high=200.0, size=nobs)
noise=np.random.normal(loc=0.0, scale=5.0, size=nobs)

yobs=xobs**2+3*xobs+7+noise
x=xobs.reshape(-1,1)
y=yobs

lr=LinearRegression()
lr.fit(x,y)

print('The coefficient = ',lr.coef_)
print('The intercept =',lr.intercept_)

fig=plt.figure(figsize=(21,8))
fig.suptitle('Machine Learning Linear Regression in Supervised Learning - Regression Algorithm.',color='m',fontsize=16)
plt.plot(xobs,yobs,'o')
xgrid=np.linspace(xobs.min(),xobs.max(),num=100)
plt.plot(xgrid, lr.coef_[0]*xgrid+lr.intercept_)

xpredict=np.array([50,100,150,200])
ypredict1=lr.intercept_+xpredict*lr.coef_[0]
ypredict2=lr.predict(xpredict.reshape(-1,1))

print('Predictions using the linear regression supervised learning algorithm')
print('Selected x values: ',xpredict)
print('Method 1: ',ypredict1)
print('Method 2: ',ypredict2)

plt.show()
plt.savefig('t6q7.png')
```
